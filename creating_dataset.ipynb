```python
import random
import pandas as pd

ambiguous = [
    "The system should be user-friendly.",
    "The website must be fast.",
    "The interface should be modern.",
    "The software should be reliable."
]

complete = [
    "The login page shall authenticate users using email and password within 2 seconds.",
    "The system shall generate a PDF report containing user statistics each month.",
    "The app must support English and Spanish languages for all screens."
]

inconsistent = [
    "The app shall store data locally and shall not store any data locally.",
    "The user must log in before use, but login should not be required.",
    "The system should send email alerts and should never send email alerts."
]

data = []
for i in range(1, 650):
    label = random.choice(["Ambiguous", "Complete", "Inconsistent"])
    if label == "Ambiguous":
        sentence = random.choice(ambiguous)
    elif label == "Complete":
        sentence = random.choice(complete)
    else:
        sentence = random.choice(inconsistent)
    data.append([sentence, label])

df = pd.DataFrame(data, columns=["requirement_text", "quality_label"])

from nlpaug.augmenter.word import SynonymAug
aug = SynonymAug(aug_src='wordnet')
df['requirement_text'] = df['requirement_text'].apply(lambda x: aug.augment(x))
```

```python
print(df['quality_label'].value_counts())
df.to_csv("requirement_quality_dataset.csv", index=False)
print("Dataset created successfully!")
```
